## Daily-Monthly- Yearly Average Return


```python
average_daily_change=[]
average_monthly_change=[]
average_yearly_change=[]

d=[]
m=[]
y=[]

stock=list(my_data.columns)
for i in stock:
    if i=='DATE' :
        continue
    for j in range(len(stock)-1):
        price_change=my_data[i][j+1]-my_data[i][j]
        rate_daily_change=(price_change/my_data[i][j])*100
        d.append(rate_daily_change)
    for k in range(0,len(stock)-1,30):
        price_change_monthly=my_data[i][k+30]-my_data[i][k]
        rate_monthly_change=(price_change_monthly/my_data[i][k])*100
        m.append(rate_monthly_change)
    for l in range(0,len(stock)-1,365):
        price_change_yearly=my_data[i][l+365]-my_data[i][l]
        rate_yearly_change=(price_change_yearly/my_data[i][l])*100
        y.append(rate_yearly_change)
        
        
    average_daily_change.append(sum(d)/len(d))
    average_monthly_change.append(sum(m)/len(m))
    average_yearly_change.append(sum(y)/len(y))
    d=[]
    m=[]
    y=[]
    
    

        
df=pd.DataFrame()
df['Stocks']=pd.Series(stock)
df['Daily_Returns']=pd.Series(average_daily_change)
df['Monthly_Returns']=pd.Series(average_monthly_change)
df['Years_Returns']=pd.Series(average_yearly_change)
df.dropna(inplace=True)
df.drop([47],inplace=True)
df.drop([0],inplace=True)
df.Daily_Returns=round(df.Daily_Returns,2)
df.Monthly_Returns=round(df.Monthly_Returns,2)
df.Years_Returns=round(df.Years_Returns,2)
```
